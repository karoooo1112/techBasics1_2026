### Commenting

Any line beginning with `#`, the pound sign, are not treated as code. They are messages for human readers and Python will simply ignore it.

Comments are used for explaining how a program works, to both the programmer themselves, and others reading the program code.

In the code cell below, the comment explains the calculation performed in the code:

```python
print("Number of students registered in Tech Basics I:")
# there are in total 2 sessions of Tech Basics I and each has 21 students
print(2*21)
```

`#` only works for a single line of comment. If you have more to write, you can use triple quotes `'''` instead.

```python
'''
This is an example of a multi-line comment.
You can write as much text as you want between the triple quotes without worrying about line breaks.
'''

print("This is the only message that is being printed")
```
