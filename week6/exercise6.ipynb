### Review: List Methods
#### Adding and removing items from a list
You can add new items to the end of a list by using `.append()`. It takes only one parameter at a time. Similar as a string, you can use `len()` to get the current length of a list.

```python
numbers = [1, 2, 3, 4, 5]

print(len(numbers))

numbers.append(7)
numbers.append(6)

print(numbers)
print(len(numbers))
```

You can also add new items to a specific location by using `.insert()`. The first parameter takes the index, while the second one takes the value.

```python
numbers = [1, 2, 3, 4, 5]
numbers.insert(0,0) # index, value
print(numbers)
numbers.insert(2,10)
print(numbers)
```

There are two ways to remove an item from a list:
- If the index of the item is known, you can use the method `.pop()`.
- If the content of the item is known, you can use the method `.remove()`.

```python
numbers = [1, 2, 3, 4, 5]

print(numbers.pop(0))
print(numbers)
numbers.pop() # if no index is given, pop removes the last item of a list
print(numbers)
```

Notice that `.pop()` not only removes the item at a given location, it also returns it. This can be very useful in some situations.

```python
list_of_fruits = ["apple", "banana", "orange"]

for i in range(3):
    print("I ate " + list_of_fruits.pop())

print(list_of_fruits)
```

The method `.remove()` on the other hand, takes the value of the item that needs to be removed as its argument.

```python
list_of_fruits = ["apple", "banana", "orange"]
list_of_fruits.remove("banana")
print(list_of_fruits)
```

If there are multiple items with the same value in a list,  it removes the first occurrence of the given value.

```python
list_of_fruits = ["apple", "banana", "orange", "banana"]
list_of_fruits.remove("banana")
print(list_of_fruits)
list_of_fruits.remove("banana")
print(list_of_fruits)
```

### Changing order of a list
The items in a list can be sorted by using the method `sort()` or the function `sorted()`. If you want to sort the list descending, add an extra parameter`reverse=True`. By default, reverse is set to False.

```python
my_list = [1, 2, 7, 6, 3, 8]
my_list.sort()
print(my_list)
my_list.sort(reverse=True)
print(my_list)
my_list = ["Anna", "Jan","Cathrine", "Julia", "Pauline"]
print(sorted(my_list))
print(sorted(my_list, reverse=True))
```

In the example above, the `.sort()` method modifies the list itself. Sometimes we don't want to change the original list, in that case we can use the function `sorted()` instead. It takes a list as a parameter and returns a new sorted list.

```python
my_list = [1, 2, 7, 6, 3, 8]

sorted_list = sorted(my_list)

print(sorted_list)
print(my_list)
```

`.sort()` and `sorted()` can be used to sort both numbers and strings in alphabetical order.

```python
my_list = ["Anna", "Jan","Cathrine", "Julia", "Pauline"]
print(sorted(my_list))
print(my_list)
my_list.sort()
print(my_list)
```

We can also use the random library to shuffle a list.

```python
import random

ingredients = ["milk","sugar","butter","egg"]
random.shuffle(ingredients)
print(ingredients)
```

### Reference

A reference tells us where the value can be found. The function `id()` can be used to find out the exact location the variable points to.
The ID of the variable, is an integer, which can be thought of as the address in computer memory where the value of the variable is stored. If you execute the following code on different computers, the result will likely be different.

```python
# Integers
# Integer is immutable
# number -> address:4315545840 -> 1
# number -> address:4315546160 -> 11
# a -> address: 4315545840 -> 1
number = 1
print(id(number))
number += 10
print(id(number))
a = 1 # See how the value 1 points to the same location of the memory
print(id(a))
print("---")

# Strings
my_text = "This is a reference, too"
print(id(my_text))
my_text = my_text + "!"
print(id(my_text))
print("---")

# List and other collections
# List is mutable
# my_list -> the same location inside your memory
my_list = [1,2,3]
print(id(my_list))
my_list[0] = 10
print(id(my_list))
print("---")

my_dictionary = {"a":1,"b":2,"c":3}
print(id(my_dictionary))
my_set = {"a","b","c"}
print(id(my_set))
print("---")

my_tuple = (1,2,3)
print(id(my_tuple))
my_tuple = (1,3,5)
print(id(my_tuple))
```

More than one reference to the same list

```python
list1 = [1, 2, 3, 4]
list2 = list1 # we are not passing the values being stored in the list, but the address

print(id(list1), id(list2)) # same id

# a change made through any one of the references affects also the other references, as their target is the same.
list1[0] = 10
list2[1] = 20

print(list1, list2)
```

That's why if you want to create an actual separate copy of a list without pointing to the same target, better use `.copy()`

```python
list3 = list1.copy()
print(id(list1),id(list3))

list3[1] = 30
print(list1, list3)
```

If you pass a list as an argument to a function, even if it's a local variable, you are still able to edit it inside a function, see the following comparison between `edit_list()` and `edit_int()`

```python
def edit_list(l:list):
    l[1] = 40

def edit_int(n:int):
    n = 5

def test_list_reference():
    test_list = [1,3,9]
    test_int = 3
    edit_list(test_list)
    edit_int(test_int)
    print(test_list)
    print(test_int)

test_list_reference()
```

However, if you create a new list inside a function and try to pass it to a list passed through parameter, it won't work because this new list is not accessible from outside the function.

```python
def augment_all(l: list):
    new_list = [] # -> local memory
    for item in l:
        new_list.append(item + 10)
    l = new_list

numbers = [1, 2, 3]
print("in the beginning:", numbers)
augment_all(numbers)
print("after the function is executed:", numbers)
```

### Dictionary

A dictionary can consist of heterogeneous data types.

```python
my_dictionary = {
    "user_name": "xyz",
    "age": 22,
    "hp": 100,
    "mp": 50,
    "tools": ["poison", "sword","key"],
    "game_over": False
}
```

Both using the index key directly as well as using `.get()` can returns the value of the item with the specified key. However, using `.get()` can avoid error messages if you are not sure if the key exist or not.

```python
# try to get a value that doesn't exist
print(my_dictionary.get("dream"))

# if "dream" in my_dictionary:
#     print(my_dictionary["dream"]) # key error if the key doesn't exist
```

You can also get a list of only keys, only values, or a list of key value pairs as tuples.

```python
print(my_dictionary.keys())
print(my_dictionary.values())
print(my_dictionary.items())
```

The familiar `for item in collection` loop can be used to traverse a dictionary too. When used on the dictionary directly, the loop goes through the keys stored in the dictionary, one by one.

```python
for key in my_dictionary:
    print(key, my_dictionary[key])
```

You can also traverse the entire  dictionary with .items()

```python
for (key, value) in my_dictionary.items():
    print(key, value)
```

You can update a dictionary just by using the key index or use the `.update()` method.

```python
my_dictionary["age"] = 23 # one value at a time
my_dictionary.update({"dream":"Travel to Antarctica",
                      "role":"magician"}) # update can accept multiple items
print(my_dictionary)
```

You can remove a specific item from a dictionary by using `.pop()` or the last inserted one with `.popitem()`

```python
print(my_dictionary.pop("dream")) # it returns the value
print(my_dictionary)
my_dictionary.popitem()
print(my_dictionary)
```

To check if a key already exist in a dictionary, use the membership operator `in`:

```python
if "dream" in my_dictionary:
    print(my_dictionary["dream"])
else:
    print("no dream :(")
```

Similar as what we did to a list, a dictionary can also be updated inside a function:

```python
def scene1():
    my_dictionary["tools"].append("torch")
    my_dictionary["hp"] -= 20

scene1()
print(my_dictionary["tools"])
print(my_dictionary["hp"])
```

### Tuple

The following code creates a tuple containing the coordinates of a point. The items stored in a tuple can be accessed by index, just like the items in a list.

```python
point = (5,8)
print("x coordinate:", point[0])
print("y coordinate:", point[1])
```

The values stored in a tuple cannot be changed after the tuple is initiated.  The following code will throw an error:

```python
point[0] = 15
```

The parentheses are not strictly necessary when defining tuples. The following two variable assignments are identical in their results:

```python
numbers = (1,2,3,4,5)
numbers = 1,2,3,4,5
 # same logic of indexing applies to tuple
print(numbers[-1])
print(numbers[2:])
```

Since tuples are immutable, they do not have a built-in `append()` or `remove()` method, but there are other ways to add/remove items to a tuple.

```python
thistuple = ("apple", "banana", "cherry")
# we can convert a tuple to a list, so that we can change its items
y = list(thistuple)
y.append("orange")
y.remove("apple")
thistuple = tuple(y)
print(thistuple)
```

When creating a tuple with only one item, remember to include a comma after the item, otherwise it will not be identified as a tuple.

```python
single_tuple = 1,
print(type(single_tuple))
single_tuple = (1)
print(type(single_tuple))
```

You can count the number of times a value appears in a tuple with `.count()`:

```python
this_tuple = (1, 3, 7, 8, 7, 5, 4, 6, 8, 5)

x = this_tuple.count(5)

print(x)
```

Or search the first occurrence of a value and get its position with `.index()`

```python
y = this_tuple.index(7)
print(y)
```

You can use tuples as keys for dictionary, because they are immutable

```python
tuple_dictionary = {}

tuple_dictionary[(0,0)] = "upper left "
tuple_dictionary[(1,1)] = "  center   "
tuple_dictionary[(2,2)] = "lower right"
```

#### Set
A set is a collection of different items. The values `True` - `1` and `False` - `0` are considered the same value in sets, and are treated as duplicates


```python
unique_set = {True, 1,0, False} # if you init a set with duplicates, the later ones won't be inserted
print(unique_set)
```

Since a set is not ordered, to add an item is called `.add()` instead of `.append()`, or you can always use `.update()` for multiple items.

```python
integer_set = {1,2,3,4,5}
integer_set.add(6)
print(integer_set)
integer_set.update({7,8})
print(integer_set)
```

To remove an item you have two choices: `.remove()` or `.discard()`. The `remove()` method will raise an error if the specified item does not exist, while the `discard()` method will not.

```python
integer_set.remove(6)
print(integer_set)
integer_set.discard(6)
print(integer_set)
```
