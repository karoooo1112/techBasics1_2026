### Function

In Python we define a function with the syntax `def`

```python

def my_function():
    pass
# this function doesn't take any parameter
    # If there is no content inside a function yet, at least write pass to avoid error messages

# def sum(a, b): # you can take multiple parameters
#     return a + b # return a value

def list_items(food):
    for item in food:
        print(item)

print(sum(3,5))
print(sum(3+21,10+7)) # you can pass evaluation as arguments
list_items(["juice","yogurt","bread"])  # you can pass a list as an argument
another_list = ["desk","chair","bed"]
list_items(another_list) # or a variable
```

If you declare a variable inside a function, it is considered as a <i> local variable </i>. It is not accessible from outside.

```python
def my_function():
    # inside my_function
    my_unique_variable = 10
    print("my_variable", my_variable)

# global, outside of function
print(my_unique_variable)
```

Compare the code above with the following code:

```python
def my_function():
    # scene 1, dragon killed me
    global lives_left
    lives_left -= 1

lives_left = 3 # global
print(lives_left)
my_function()
print(lives_left)
```

What happens here is that `my_function()` assigns a value to `my_varialbe`. Inside the function, Python interprets `my_varialbe` to be a local variable instead of the global variable of the same name. The function attempts to access the variable before it is defined, so there is an error.

We can fix this error by changing the order of printing inside the function, the value of a global variable won't be changed from within another function. See:

```python
my_variable = 5

def my_function():
    my_variable = 10
    print("inside the function:", my_variable)

my_function()
print("outside the function:", my_variable) # the local variable assignment has zero effect on the global variable
```

You can access any global variables inside a function without using any specific syntax. However, if you really wish to change the value of a global variable inside a function, you have to add the Python keyword `global`. But this in general is not considered a good practice and one shall avoid using it if absolute necessary.

The proper way to do this would be passing the global variable to the function as an argument, then return the new value. In this way, we are making our code easier to maintain and debug as things grow more complicated.

```python
def access_demo():
    # you don't need keyword global just to access global variable
    print(f"Access Demo: x={x}")

def local_demo():
    x = 1 # by default this is local
    print(f"Local Demo: x={x}")

def global_demo():
    global x # the global keyword hack
    x = 3 # here python won't interpret x anymore as a local variable
    print(f"Global Demo: x={x}")

def update_number(number): # proper way to do this
    return number + 2

x = 5 # global variable
access_demo()
local_demo()
print(f"Outside: x={x}")
global_demo()
print(f"Outside: x={x}")
x = update_number(x)
print(f"Outside: x={x}")
```

#### Positional Arguments

In Python, a function can take arguments either based on its position or based on its name.




```python
def print_arguments(arg1, arg2):
    print(arg1, arg2, arg1-arg2)

# position matters
print_arguments(1, 2)
print_arguments(2, 1)
```

#### Arbitrary Arguments, *args

If you want to be flexible with how many arguments that can be passed into your function, add a * before the parameter name in the function definition.

This way the function can receive one or multiple arguments.

```python
def show_classes(*classes):
    # print("The first class is:", classes[0])
    for c in classes:
        print(c)

show_classes("Agents and Interface")
show_classes("Technical Basics I", "Technical Basics II")
```

#### Keyword Arguments

You can also send arguments with the `key = value` syntax. This way the order of the arguments does not matter any more, but the keywords.

```python
def my_data(name, birthday, email):
    print("Name:", name, "Birthday:", birthday, "Email:", email)

my_data( name = "Max Mustermann", email="max.mustermann@leuphana.de", birthday = "01.01.2000")

my_data( "Max Mustermann", "max.mustermann@leuphana.de", "01.01.2000")
```

#### Arbitrary Keyword Arguments, **kwargs

If you do not know how many keyword arguments that will be passed into your function, add two asterisk: ** before the parameter name in the function definition.

This way the function will receive a *dictionary* of arguments. We will talk more about dictionary as a data type next week.

```python
def my_data(**data):
    for key in data:
        print(key, data[key])

my_data( name = "Max Mustermann", email="max.mustermann@leuphana.de", birthday = "01.01.2000", major = "Digital Media")
```

#### Default Parameter Value

You can also add a default parameter value to your function. If the function is called without argument, it uses the default value.

```python
def my_status(status = "normal"):
    print("Status:", status)

my_status("good")
my_status("bad")
my_status()
```

#### Type hints

Sometimes our functions are very sensitive to the type of the arguments provided. Take a look at the following function:

```python
def print_many_times(message, times):
    while times > 0:
        print(message)
        times -= 1
```

It works as expected if we provide the values that it assumes:

```python
print_many_times("Hello World!", 3)
```

However, in some cases, it won't work if we give an argument of the wrong type.

```python
print_many_times("Hello World", "4")
```

To avoid issues like this you can always check the parameter types inside your function. And to make it clear for everyone who wants to use your function, you can add

```python
def print_many_times(message : str, times : int):
    while times > 0:
        print(message)
        times -= 1
```

#### Task 3: Passing data from one function to another

Finish the following code and observer how data has been passed around to avoid using global variables.

```python
def input_from_user(how_many:int):
    # this function gets input from the user, it takes one integer parameter, and it defines how many inputs the program asks. It shall return the input numbers as a list of integer
    list_of_integer = [] # start with an empty list
    for i in range(how_many):
        n = int(input("Please enter a number"))
        list_of_integer.append(n)
    return list_of_integer

def print_result(numbers:list):
    # this function displays all the numbers and the total sum
    print("Input:", numbers)
    print("Total Sum:", sum(numbers))

def main():
    inputs = input_from_user(3)
    print_result(inputs)

main()
```

### Recursion

We can use recursion wisely to achieve similar results as a loop


```python
def count_down(n):
    if n == 0: # a recursive function always needs an end point to prevent from forever looping
        return
    print(n)
    count_down(n - 1)
    # try to move print(n) to here and see what happens?

# Example usage:
count_down(5)
```

It can also do unique things such as branching:

```python
def branch_print(n, depth=0):
    if n == 0:
        return
    print('  ' * depth + f"Node {n}")
    branch_print(n - 1, depth + 1)  # First branch (deeper)
    branch_print(n - 1, depth + 1)  # Second branch (deeper)

# Example usage:
branch_print(3)
```
