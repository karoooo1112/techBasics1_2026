### Task 2

```python
def custom_greeting(name="dear guest", greeting="Hello"):
    message = greeting + ", " + name + "!"
    return message

print(custom_greeting(name="Bob", greeting="Welcome"))  # Output: "Welcome, Bob!"
print(custom_greeting(name="Alice"))  # Output: "Hello, Alice!"
print(custom_greeting(greeting="Good evening"))  # Output: "Good evening, dear guest!"
```

### Task 3

```python
def input_from_user(how_many:int):
    result = [] # start an empty list
    for i in range(how_many):
        result.append(int(input("Please give me a number"))) # raw
    return result

def print_result(numbers:list):
    print("Inputs:", end=" ")
    print(numbers)
    print("Total Sum: ", sum(numbers))

def main():
    inputs = input_from_user(3)
    print_result(inputs)

main()
```

### Task 4

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

factorial(4) # 4 * 3 * 2 * 1
```
